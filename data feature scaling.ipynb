## Missing values

```python
data={'A':[1,2,4,5,6],'B':[10,40,np.nan,80,60],'C':[np.nan,np.nan,100,50,40]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
from sklearn.impute import SimpleImputer
```

```python
sm=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
sm.fit(df)
```

```python
sm.transform(df)
```

```python
sm1=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
sm1.fit(df)
```

```python
sm1.transform(df)
```
